```python
import csv

# Define input and output file paths
input_file_path = r"C:\Users\marga\Downloads\iCysMod_all_data\iCysMod_all_data.txt"
output_file_path = r"C:\Users\marga\Downloads\iCysMod_all_data\iCysMod_all_data.csv"

# Read tab-separated txt file and write as CSV file
with open(input_file_path, 'r', newline='', encoding='utf-8') as txt_file, \
     open(output_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    txt_reader = csv.reader(txt_file, delimiter='\t')
    csv_writer = csv.writer(csv_file)

    for row in txt_reader:
        csv_writer.writerow(row)
```

```python
#Glutathione
import pandas as pd

# Read the two CSV files into Pandas DataFrames
df1 = pd.read_csv(r"C:\Users\marga\Downloads\iCysMod_all_data\filtered_data.csv")
df2 = pd.read_csv(r"C:\Users\marga\OneDrive\Desktop\summer project\glu_byonic.csv")

# Subtract 1 from the 'position' column in df2
df2['position'] = df2['position'] - 1

# Create sets of unique (protein_id, position) combinations for each dataframe
set_df1 = set(zip(df1['protein_id'], df1['position']))
set_df2 = set(zip(df2['protein_id'], df2['position']))

# Find combinations only in df1
only_in_df1 = set_df1 - set_df2
only_in_df1_length = len(only_in_df1)

# Find combinations only in df2
only_in_df2 = set_df2 - set_df1
only_in_df2_length = len(only_in_df2)

# Find combinations in both df1 and df2
in_both = set_df1 & set_df2
in_both_length = len(in_both)

# Convert the results back to DataFrames
only_in_df1_df = pd.DataFrame(only_in_df1, columns=['protein_id', 'position'])
only_in_df2_df = pd.DataFrame(only_in_df2, columns=['protein_id', 'position'])
in_both_df = pd.DataFrame(in_both, columns=['protein_id', 'position'])

# Print the results
print("Combinations only in icysmod:")
print(f"Number of combinations: {only_in_df1_length}")
#print(only_in_df1_df)

print("\nCombinations only in byonic data:")
print(f"Number of combinations: {only_in_df2_length}")
#print(only_in_df2_df)

print("\nCombinations in both df1 and df2:")
print(f"Number of combinations: {in_both_length}")
#print(in_both_df)
```
